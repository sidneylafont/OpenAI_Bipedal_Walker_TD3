```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#reading in results
training_50_ep = pd.read_csv(f'results/third_training_attempt/train_50_episodes_results.csv')
training_100_ep = pd.read_csv(f'results/third_training_attempt/train_100_episodes_results.csv')
training_150_ep = pd.read_csv(f'results/third_training_attempt/train_150_episodes_results.csv')
training_200_ep = pd.read_csv(f'results/third_training_attempt/train_200_episodes_results.csv')
training_250_ep = pd.read_csv(f'results/third_training_attempt/train_250_episodes_results.csv')
training_300_ep = pd.read_csv(f'results/third_training_attempt/train_300_episodes_results.csv')
training_350_ep = pd.read_csv(f'results/third_training_attempt/train_350_episodes_results.csv')
training_400_ep = pd.read_csv(f'results/third_training_attempt/train_400_episodes_results.csv')
training_450_ep = pd.read_csv(f'results/third_training_attempt/train_450_episodes_results.csv')
training_500_ep = pd.read_csv(f'results/third_training_attempt/train_500_episodes_results.csv')
training_550_ep = pd.read_csv(f'results/third_training_attempt/train_550_episodes_results.csv')
training_600_ep = pd.read_csv(f'results/third_training_attempt/train_600_episodes_results.csv')
training_610_ep = pd.read_csv(f'results/third_training_attempt/train_610_episodes_results.csv')
```

```python
df = training_50_ep.append(training_100_ep)
df = df.append(training_150_ep)
df = df.append(training_200_ep)
df = df.append(training_250_ep)
df = df.append(training_300_ep)
df = df.append(training_350_ep)
df = df.append(training_400_ep)
df = df.append(training_450_ep)
df = df.append(training_500_ep)
df = df.append(training_550_ep)
df = df.append(training_600_ep)
df = df.append(training_610_ep)
df = df.drop(['Unnamed: 0'], axis=1)
df['episode'] = list(range(len(df)))
df = df.rename(columns={"0": "score"})
```

```python
fig = plt.figure()
ax = plt.axes()

x = df['episode'].tolist()
y = df['score'].tolist()

ax.plot(x, y)
plt.show()
```

```python
fig.savefig(f'results/third_training_attempt/rewards_graph.png')
```
